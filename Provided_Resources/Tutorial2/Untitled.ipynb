```julia
using ControlSystems
using Plots

# plotly()
```

```julia
s = tf('s')
```

```julia
# Rotor Plant
P = 1 / s^2;

## Plotting
p1 = bodeplot(P)
p2 = pzmap(P)
plot(p1, p2)
```

```julia
# PD Controller Plant
Kd = 4;
Kp = 0.5;


C = Kd*s + Kp;

## Plotting
p1 = bodeplot(C)
p2 = pzmap(C)
plot(p1, p2)
```

```julia
# Root Locus of the Feedback Transfer Function
rlocusplot(C*P, xlim=[-0.5,0.1], ylim=[-0.5, 0.5])
```

```julia
C*P
```

```julia
# Sensitivity Function
S = 1 / (1 + C*P)

## Plotting
p1 = bodeplot(S)
p2 = pzmap(S)
display(plot(p1, p2))
S
```

```julia
# Complimentary Sensitivity Function
CS = C * P * S

## Plotting
p1 = bodeplot(CS)
p2 = pzmap(CS)
display(plot(p1, p2))
CS
```
